```python
class Queens:
    def __init__(self, num):    # constructor
        self.num = num
        self.board = [[0]*num for _ in range(num)]
        self.solutions = []
        self.solution_count = 0

    def safe_place(self, row, col):   
        for i in range(row):            # search from left row
            if self.board[i][col] == 1:
                return False
            
        r, c = row, col             # search upper left diagonal
        while r >= 0 and c >= 0:
            if self.board[r][c] == 1:
                return False
            r -= 1
            c -= 1
        
        r, c = row, col               # search upper right diagonal
        while r >= 0 and c < self.num:
            if self.board[r][c] == 1:
                return False
            r -= 1
            c += 1
        
        return True
        
    def placing_queen(self, row=0):
        if row == self.num:
            self.solution_count += 1
            self.solutions.append([row[:] for row in self.board])
            if self.solution_count == 5:
                self.print_solutions()
                return False
            
        for col in range(self.num):
            if self.safe_place(row, col):
                self.board[row][col] = 1
                if self.placing_queen(row + 1):
                    return True
                self.board[row][col] = 0
        return False
            
    def print_solutions(self):
        for i, solution in enumerate(self.solutions):
            print(f"{self.num}x{self.num} Queen's Board solution {i+1}:")
            for row in solution:
                print(' '.join('Q' if x == 1 else '*' for x in row))
            print()

queens_prob = Queens(16)  
queens_prob.placing_queen()
```
